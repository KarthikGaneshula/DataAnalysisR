#  Strategy 1 without Dividends

The following is the code for the first strategy by which we should end up with a profit by the end of the year because we are
buying low stocks.

```r

Total = 5000000
options(digits=9)
Remaining = Total
split = Total/5
MTM = 0
start_date = mdy('01/02/2018')
begin_date <- mdy('01/02/2018')
end_date =  mdy('12/28/2018')
AllStocks = c('IBM_AdjClose','MSFT_AdjClose','Google_AdjClose','Apple_AdjClose','Amazon_AdjClose','FB_AdjClose','NFLX_AdjClose','TSLA_AdjClose','ORCL_AdjClose','SAP_AdjClose')
IntialStocks = c('IBM_Close','MSFT_Close','Google_Close','Amazon_Close','FB_Close')
SharesCount = list()


while(start_date <= end_date){
    # executes the first day
    if(start_date == '2018-01-02'){
          for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        b= floor(split/a[i])
        SharesCount[i] =  b[[i]] 
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        colnames(Remaining) = 'Remaining'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        colnames(MTM) = 'MTM'
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
        
    }

    

    else if(start_date %in% StockSplit_5$Date) {

        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        Total = MTM + Remaining
        Remaining = Total
        split = Total / 5
        colnames(Total) = 'Total'
        colnames(MTM) = 'MTM'
        calculated_stocks = list()
        for(i in AllStocks){
           
            x = Stocks[Stocks$Date== start_date, AllStocks]
            y = Stocks[Stocks$Date== begin_date, AllStocks]
            percent_change = 100 * ((x[i] - y[i])/ y[i])
            col = paste0(i, "_PC")
            calculated_stocks[col] = percent_change
            
        }
        calculated_stocks = calculated_stocks[order(unlist(calculated_stocks))]
        calculated_stocks = head(calculated_stocks,5)
        

        initial_col_list = names(calculated_stocks)
        print(initial_col_list)
        initial_col_list = sub('_.*','',initial_col_list)
        initial_col_list = paste0(initial_col_list,'_Close')
        IntialStocks = initial_col_list
        SharesCount = list()
        b = list()
        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        b= floor(split[,1]/a[i])
        SharesCount[i] =  b[[i]]
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        
        
    }
        

    else {
        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        colnames(MTM) = 'MTM'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
      
    }
start_date = start_date+1

}
sr1_5 = c('StockSplit_5',MTM = tail(Stocks$MTM,1))
output_1 = c('Profit:', MTM-5000000)
sr1_5
output_1
tail(Stocks)
```

# Strategy-1, every 8 days

```r
StockSplit_8 = Stocks[seq(1,nrow(Stocks),8),]
StockSplit_8$Date = as.Date(StockSplit_8$Date)
head(StockSplit_8)
```

```r
Total = 5000000
options(digits=9)
Remaining = Total
split = Total/5
MTM = 0
start_date = mdy('01/02/2018')
begin_date <- mdy('01/02/2018')
end_date =  mdy('12/28/2018')
AllStocks = c('IBM_AdjClose','MSFT_AdjClose','Google_AdjClose','Apple_AdjClose','Amazon_AdjClose','FB_AdjClose','NFLX_AdjClose','TSLA_AdjClose','ORCL_AdjClose','SAP_AdjClose')
IntialStocks = c('IBM_Close','MSFT_Close','Google_Close','Amazon_Close','FB_Close')
SharesCount = list()

while(start_date <= end_date){
    if(start_date == '2018-01-02'){
          for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        b= floor(split/a[i])
        SharesCount[i] =  b[[i]] 
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        colnames(Remaining) = 'Remaining'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        colnames(MTM) = 'MTM'
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
        
    }

    
    else if(start_date %in% StockSplit_8$Date) {
        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        Total = MTM + Remaining
        Remaining = Total
        split = Total / 5
        colnames(Total) = 'Total'
        colnames(MTM) = 'MTM'
        calculated_stocks = list()
        for(i in AllStocks){
           
            x = Stocks[Stocks$Date== start_date, AllStocks]
            y = Stocks[Stocks$Date== begin_date, AllStocks]
            percent_change = 100 * ((x[i] - y[i])/ y[i])
            col = paste0(i, "_PC")
            calculated_stocks[col] = percent_change
            
        }
        calculated_stocks = calculated_stocks[order(unlist(calculated_stocks))]
        calculated_stocks = head(calculated_stocks,5)
        
    
        initial_col_list = names(calculated_stocks)
        print(initial_col_list)
        initial_col_list = sub('_.*','',initial_col_list)
        initial_col_list = paste0(initial_col_list,'_Close')
        IntialStocks = initial_col_list
        SharesCount = list()
        b = list()
        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        b= floor(split[,1]/a[i])
        SharesCount[i] =  b[[i]]
        Remaining = Remaining - b*a[i]
        MTM = MTM + b*a[i]
        }
        MTM = MTM + Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        
        
    }
        

    else {
        MTM = 0
        for(i in IntialStocks) {
        a = Stocks[Stocks$Date== start_date, IntialStocks]
        MTM = MTM + SharesCount[[i]]*a[i]
        }
        colnames(MTM) = 'MTM'
        Stocks[Stocks$Date == start_date,'Remaining'] = Remaining
        Stocks[Stocks$Date == start_date,'MTM'] = MTM
      
    }
start_date = start_date+1
}
sr1_8 = c('StockSplit_8',MTM = tail(Stocks$MTM,1))
output_5 = c('Profit:', MTM-5000000)
sr1_8
output_5
tail(Stocks)
```
